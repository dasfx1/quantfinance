```python
# üìç Notebook-Zelle 1
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")
%matplotlib inline
```

```python
# üìç Notebook-Zelle 2
df = pd.read_csv("../results/backtest_example.csv", index_col=0, parse_dates=True)
df.head()
```

```python
# üìç Notebook-Zelle 3
plt.figure(figsize=(12, 4))
plt.plot(df.index, df["final_capital"], label="Kapital")
plt.title("Kapitalverlauf ‚Äì Backtest")
plt.xlabel("Datum")
plt.ylabel("Kontostand")
plt.legend()
plt.tight_layout()
plt.show()
```
